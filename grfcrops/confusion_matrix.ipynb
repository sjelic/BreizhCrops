```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np
```

```python
# X, y = make_classification(random_state=0)
# X_train, X_test, y_train, y_test = train_test_split(
#         X, y, random_state=0)
# clf = SVC(random_state=0)
# clf.fit(X_train, y_train)
fold = 1
country = 'serbia'
model = 'TransformerPretrainedIN'
y_true = np.load(f'../results/{country}/{fold}/{model}/y_true.npy')
y_pred= np.load(f'../results/{country}/{fold}/{model}/y_pred.npy')
normalize = 'true'
# y_pred = clf.predict(X_test)
ConfusionMatrixDisplay.from_predictions(
   y_true = y_true, y_pred = y_pred, display_labels=[
            'Barley',
            'Maize',
            'Rapeseed',
            'Soya',
            'Sugarbeet',
            'Sunflower',
            'Wheat'
], xticks_rotation = 'vertical', cmap=plt.cm.Blues, normalize = normalize)
plt.savefig(f"./images/{model}_{country}_{fold}_{normalize}.pdf",bbox_inches='tight')
plt.show()
```
